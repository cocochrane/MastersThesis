## Code to Import Python Notebook Classes

```python
import io, os, sys, types
import pandas as pd
from IPython import get_ipython
from nbformat import current
from IPython.core.interactiveshell import InteractiveShell

def find_notebook(fullname, path=None):
    """find a notebook, given its fully qualified name and an optional path

    This turns "foo.bar" into "foo/bar.ipynb"
    and tries turning "Foo_Bar" into "Foo Bar" if Foo_Bar
    does not exist.
    """
    name = fullname.rsplit('.', 1)[-1]
    if not path:
        path = ['']
    for d in path:
        nb_path = os.path.join(d, name + ".ipynb")
        if os.path.isfile(nb_path):
            return nb_path
        # let import Notebook_Name find "Notebook Name.ipynb"
        nb_path = nb_path.replace("_", " ")
        if os.path.isfile(nb_path):
            return nb_path


class NotebookLoader(object):
    """Module Loader for Jupyter Notebooks"""
    def __init__(self, path=None):
        self.shell = InteractiveShell.instance()
        self.path = path

    def load_module(self, fullname):
        """import a notebook as a module"""
        path = find_notebook(fullname, self.path)

        print ("importing Jupyter notebook from %s" % path)

        # load the notebook object
        with io.open(path, 'r', encoding='utf-8') as f:
            nb = current.read(f, 'json')


        # create the module and add it to sys.modules
        # if name in sys.modules:
        #    return sys.modules[name]
        mod = types.ModuleType(fullname)
        mod.__file__ = path
        mod.__loader__ = self
        mod.__dict__['get_ipython'] = get_ipython
        sys.modules[fullname] = mod

        # extra work to ensure that magics that would affect the user_ns
        # actually affect the notebook module's ns
        save_user_ns = self.shell.user_ns
        self.shell.user_ns = mod.__dict__

        try:
            for cell in nb.worksheets[0].cells:
                if cell.cell_type == 'code' and cell.language == 'python':
                    # transform the input to executable Python
                    code = self.shell.input_transformer_manager.transform_cell(cell.input)
                    # run the code in themodule
                    exec(code, mod.__dict__)
        finally:
            self.shell.user_ns = save_user_ns
        return mod


class NotebookFinder(object):
    """Module finder that locates Jupyter Notebooks"""
    def __init__(self):
        self.loaders = {}

    def find_module(self, fullname, path=None):
        nb_path = find_notebook(fullname, path)
        if not nb_path:
            return

        key = path
        if path:
            # lists aren't hashable
            key = os.path.sep.join(path)

        if key not in self.loaders:
            self.loaders[key] = NotebookLoader(path)
        return self.loaders[key]

sys.meta_path.append(NotebookFinder())
```

# 2. Collect Data

### Class Initialization
The class takes multiple parameters:  
1. participant_list: the participant_list attribute of a Participants class object
2. sleep_info: dictionary that specifies the sleep information to use  
    - nPreviousNights: number of previous nights of sleep information to use
    - sleepFeatures: the features from the SleepInformation.csv file to use
3. test_info: dictionary that specifies the test information to use  
    - nPreviousTimesteps: specifies the number of previous test scores to use
    - previousTestFeatures: specific test/feature combinations to use as a list of lists. For example, [['PVT','MeanInverseRT'],['DSST','CORRECT']] includes the MeanInverseRT feature from the PVT test and the CORRECT feature from the DSST test.
    - timingInfoPreviousTests: list of timing features to include for each of the previous tests. The options are HoursAwake, NumWPonFD, CircadianPhase, NumWeekOnFD, SleepOpp72, HoursOnProtocol, SESSION, or WP. 
    - timeInfoCurrentTest: list specifies the timing features to use for the test being predicted. Same options as above.
    - ignoreFirstN: boolean determining whether to try to predict test scores where there are not 'nPreviousTimesteps' number of previous tests present in the same wake period to use as features. 
4. demographic_features: list of demographic features to use (options are given in SubjectInformation.csv)
5. output_info: dictionary of information about outcome metric  
    - outputDataType: type of data to use for outcome metric (PVT, DSST, ADD, or Moods)
    - outputFeature: feature in the 'outputDataType' file to use as the outcome metric
    - log_output: boolean whether to log the outcome metric
6. pre_processing_info: dictionary that specifies pre-processing information
    - normalize: boolean determining whether to normalize data
    - imputationType: type of column-wise imputation. Options are 'drop' (drop missing data), or impute with 'mean', 'median', or 'mode'
    - colsToNotNormalize: list of columns that should not be normalized

### Data Collection
Once the class is instantiated, the data is collected using the getData() function of the SlidingWindowDataCollection class. This function returns the collected dataframe as well as saving it to the Datasets folder. Note that this function, if there is already a file in the Datasets folder that used the same feature, will prompt the user about whether to overwrite the file or not. If 'yes', then re-collect data and overwrite existing file. If 'no', then use the pre-existing file as our data. It also writes a txt file containing the features used, and other parameter information for future reference.  

**Note**: For using the current code, cannot use any ADD, DSST, or Moods features as these files are not included in the repo. 

```python
sleep_info = {'nPreviousNights':1, 'sleepFeatures':['Wake','SWS']}
test_info = {'previousTestFeatures':[['PVT','MeanInverseRT']], 'timingInfoPreviousTests':['CircadianPhase','HoursAwake'], 'nPreviousTimesteps':1, 'timingInfoCurrentTest':['HoursAwake','CircadianPhase', 'HoursOnProtocol', 'SESSION', 'WP', 'SleepOpp72'], 'ignoreFirstN':True}
demographic_features = ['study', 'participantCode', 'age', 'gender']
output_info = {'outputDataType':'PVT', 'outputFeature':'MeanInverseRT', 'log_output':False}
pre_processing_info = {'normalize':True, 'imputationType':'median','colsToNotNormalize':['gender','study','WP(t)', 'DecimalTime', 'SESSION(t)', 'participantCode',output_info['outputDataType']+output_info['outputFeature']+"(t)"]}

participant_list = Participants().participant_list

slidingWindow = SlidingWindowDataCollection(participant_list,sleep_info, test_info, demographic_features, output_info, pre_processing_info)
data = slidingWindow.getData()
```

# 3. Get Nested Cross-Validation Splits

The nested cross-validations splits are made by the split_data() function of the SlidingWindowDataCollection class. The function takes the following arguments:  
1. collectedData: pandas dataframe that was collected using the getData() function
2. split_info: dictionary of information about how to split data
    - splitType: which nested cross-validation technique to use. Options are none, predictSecondHalf, populationInformedPredictSecondHalf, or populationInformedWPForwardChaining and are specified as strings
    - studyChoice: (optional) list of study names to use in the dataset- all others are excluded
    - peopleChoice: (optional) list of participant codes to use in the dataset- all others are excluded

```python
split_info = {'splitType': 'predictSecondHalf'}
training_sets, testing_sets, validation_sets = slidingWindow.split_data(data,split_info)
print("\nNumber of training/test splits:",len(training_sets))
```

# 4. LME Model

##### Initialization
Initialized with three dictionaries, model_info, feature_info and output_info.  
- model_info: dictionary {'modelType':type}
- feature_info: dictionary {'random_effect_features':[], 'fixed_effect_features':[], 'intercept':True/False, 'clusterBy':featureName}
- output_info: dictionary {'featureName': name, 'logOutput': True/False}

##### Main Methods
The method fitLME() fits the model to the specified training set and returns the results for the specified test set. The method get_results() takes multiple training/test splits (given as list of dataframes, produced from the DataCollection notebook) and returns the aggregated testing sets with their predictions. 

##### Other Methods
Methods RMSE() and MAE() return these metrics on the test set. The method getModelCoefficients() returns the summary as produced from the R lme4 package.

```python
model_info = {'modelType': 'MixedEffects', 'modelParameters':{}}
feature_info = {'intercept': True, 'clusterBy':'participantCode' , 'random_effect_features': [], 'fixed_effect_features': ['HoursAwake(t)', 'CircadianPhase(t)', 'HoursOnProtocol(t)', 'SleepOpp72(t)', 'PVTMeanInverseRT(t-1)']}

lme = LinearMixedEffectsModel(model_info, feature_info, output_info)
pred = lme.get_results(training_sets, testing_sets, validation_sets)
```

```python
print('RMSE:',lme.RMSE())
print('MAE:',lme.MAE())
print("\nModel Summary:")
lme.getModelCoefficients()
```

# 6. Ensemble Model

##### Initialization
Initialized with the training, validation, and testing sets as well as with three dictionaries, model_info, feature_info and output_info.  
- model_info: dictionary {'modelType':type, 'weight':#}
- feature_info: dictionary {'random_effect_features':[], 'fixed_effect_features':[], 'intercept':True/False, 'clusterBy':featureName}
- output_info: dictionary {'featureName': name, 'logOutput': True/False}

##### Main Methods
The method pickBestModel() uses cross-validation to choose the best feature and weight combination that minimizes validation set RMSE. The features that are tried are given by the list called featuresToTry. Once the best model is chosen, it is evaluated on the testing set. 

*Note*: pickBestModel was intended to work with one train/test/validate split. The training_sets, testing_sets, and validation_sets should be lists containing only 1 dataframe. Otherwise, the first train/test/split will be used to pick the best model.

The method get_results() takes multiple training/test splits (given as list of dataframes, produced from the DataCollection notebook) and returns the aggregated testing sets with their predictions. 

##### Other Methods
Methods RMSE() and MAE() return these metrics on the test set. The attributes merfModel and lmeModel give the class objects for the LinearMixedEffectsModel and MERFModel classes, respectively. With these class objects, all of the methods for extracting coefficients and producing convergence plots are exposed.

### Determine Best Ensemble Model With CV
